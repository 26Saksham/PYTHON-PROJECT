```python
from selenium import webdriver
import os
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager

driver=webdriver.Chrome(ChromeDriverManager().install())
user  = ['prasannapandya1812','mohitsoneja_','hemang_gawali','ramshikhre20']
message = "Hello,Good Morning!"
followerList=['robertdowneyjr','urvashirautela',]
class bot:
    def __init__(self,username,password,user ,message,inputval):
        self.inputval=inputval
        self.username=username
        self.password=password
        self.user=user 
        self.message=message
        self.baseurl='https://www.instagram.com/'
        self.bot=driver
        
        self.login()
    
    def login(self):
        self.bot.get(self.baseurl)
        
        #enter the username and password
        enter_username=WebDriverWait(self.bot,20).until(expected_conditions.presence_of_element_located((By.NAME, 'username')))
        enter_password=WebDriverWait(self.bot,20).until(expected_conditions.presence_of_element_located((By.NAME,'password')))
        enter_username.send_keys(self.username)
        enter_password.send_keys(self.password)
        enter_password.send_keys(Keys.ENTER)
        time.sleep(3)
        a=WebDriverWait(self.bot,20).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[2]/section/main/div/div/div/section/div/button')))
        a.send_keys(Keys.ENTER)
        b=WebDriverWait(self.bot,20).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[3]/button[2]')))
        b.send_keys(Keys.ENTER)
        if(self.inputval==1):
            self.MessageSend(self.user)
#             self.logout()
        elif(self.inputval==2 or self.inputval==3):
            self.followRequestSend()
#             self.logout()
        

        


    def MessageSend(self,user):
        clickOnMessageButton=WebDriverWait(self.bot,15).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div/div/div[2]/div[4]/div/a')))
        clickOnMessageButton.send_keys(Keys.ENTER)
 
        for i in user:
            clickOnPensilIcon=WebDriverWait(self.bot,15).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div/div[2]/div/section/div/div/div/div/div[1]/div[1]/div/div[3]/button')))
            clickOnPensilIcon.send_keys(Keys.ENTER)
            searchBoxinput=WebDriverWait(self.bot,15).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div[2]/div[1]/div/div[2]/input')))
            searchBoxinput.send_keys(i)
            clickOnUser=WebDriverWait(self.bot,15).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div[2]/div[2]/div[1]/div/div[3]/button')))
            clickOnUser.send_keys(Keys.ENTER)
            
            clickOnNextButton=WebDriverWait(self.bot,15).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div[1]/div/div[3]/div/button')))
            clickOnNextButton.send_keys(Keys.ENTER)
            time.sleep(2)
            TypeMessage=WebDriverWait(self.bot,15).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div/div[2]/div/section/div/div/div/div/div[2]/div[2]/div/div[2]/div/div/div[2]/textarea')))
            TypeMessage.send_keys(message)
            time.sleep(2)
            clickOnSendButton=WebDriverWait(self.bot,15).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div/div[2]/div/section/div/div/div/div/div[2]/div[2]/div/div[2]/div/div/div[3]/button')))
            clickOnSendButton.send_keys(Keys.ENTER)
        
        clickOnHomePage=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div/div[1]/div/div/div/div[1]/div/div/a')))
        clickOnHomePage.send_keys(Keys.ENTER)       
    def ClearSearchHistory(self):
        clickOnHomePage=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div/div/div[1]/div/div/a')))
        clickOnHomePage.send_keys(Keys.ENTER) 
    
        clickOnSearchButton=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div/div/div[2]/div[2]/div/a')))
        clickOnSearchButton.send_keys(Keys.ENTER)

        clearAll=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div[2]/div/div/div[2]/div[2]/div/div/button')))
        time.sleep(2)
        clearAll.send_keys(Keys.ENTER)
        clear=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div/div/div[2]/div/div/div[2]/button[1]')))
        clear.send_keys(Keys.ENTER)
    def logout(self):
        clickOnMoreOption=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div/div[1]/div/div/div/div[3]/div/div/a')))
        clickOnMoreOption.send_keys(Keys.ENTER)
        
        logOutClick=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div/div[1]/div/div/div/div[3]/div/div/div[1]/div/div[1]/div[5]')))
        logOutClick.send_keys(Keys.ENTER)
        print("Logout....")
       
    def followRequestSend(self):
        for user in followerList:
            clickOnSearchButton=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div/div/div[2]/div[2]/div/a')))
            clickOnSearchButton.send_keys(Keys.ENTER)
            searchBox=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div[2]/div/div/div[2]/div[1]/div/input')))
            searchBox.send_keys(user)
            clickOnUser=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[1]/div[1]/div/div[2]/div/div/div[2]/div[2]/div/div[1]/div/a')))
            clickOnUser.send_keys(Keys.ENTER)
            clickOnFollowButton=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[2]/div[2]/section/main/div/header/section/div[1]/div[2]/div/div[2]/button')))
            clickOnFollowButton.send_keys(Keys.ENTER)
            time.sleep(2)
            clickOnHomePage=WebDriverWait(self.bot,10).until(expected_conditions.presence_of_element_located((By.XPATH,'/html/body/div[1]/div/div/div/div[1]/div/div/div/div[1]/div[2]/div[1]/div/div/div/div[2]/div[1]/div/div/a')))
            clickOnHomePage.send_keys(Keys.ENTER)        
        if(self.inputval==3):
            self.MessageSend(followerList)
            self.ClearSearchHistory()
            
  
  
def init():

    print('Press 1 for Send message')
    print('press 2 for send follow request')
    print('press 3 for send follow request with messgae (clear the Search History)')
    val=int(input("Enter the Input"))
    if(val  in [1,2,3]):
        bot('sakshamgupta2002', 'rajneekhra', user, message,val)
    else:
        print("Wrong press")

	# when our program ends it will show "done".
    print("DONE")
    


# calling the function
init()
```
